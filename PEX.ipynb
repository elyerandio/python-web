```python
import mechanize
from bs4 import BeautifulSoup
import re
import urlparse

pattern = re.compile('\n{2,}')             # regular expression to remove duplicate newline
URL_BASE = "http://www.pinoyexchange.com/forums/"
```

```python
def makeSoup(url):
    br = mechanize.Browser()
    br.set_handle_robots(False)
    br.addheaders = [('User-agent', 'Firefox')]
    resp = br.open(url)
    soup = BeautifulSoup(resp.get_data())
    return soup
```

```python
url = "http://www.pinoyexchange.com/forums/showthread.php?t=645717"
br = mechanize.Browser()
br.set_handle_robots(False)
br.addheaders = [('User-agent','Firefox')]
resp = br.open(url)
soup = BeautifulSoup(resp.get_data())
```

```python
post_re = re.compile('^post_')
topic = soup.find('li', attrs={'class':'navbit lastnavbit'})
posts = soup.find_all('li', attrs={'id' : post_re})
#posts = soup.find_all('blockquote', attrs={'class':'postcontent restore '})
for post in posts:
    user = post.strong.get_text()
    date = post.find('span', attrs={'class':'date'}).get_text()
    time = post.find('span', attrs={'class':'time'}).get_text()
    print '==============================================================='
    print '%s => %s' % (user, date)
    print '--------------------------------'
    msg = post.find('blockquote', attrs={'class':'postcontent restore '})
    msg = msg.get_text().strip()
    msg = pattern.sub('\n', msg)
    print msg, '\n'
    
pagination = soup.find('div', attrs={'class' : 'pagination_bottom'})
current_page = pagination.find('span', attrs={'class':'selected'})
next_page = current_page.next_sibling
print 'current page', current_page
print 'next_page', next_page
    
```

```python
next_url = urlparse.urljoin(URL_BASE, current_page.find_next_sibling().a['href'])
next_soup = makeSoup(next_url)
```
