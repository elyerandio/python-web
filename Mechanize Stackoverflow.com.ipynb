```python
from mechanize import Browser
# create a new instance of Browser class
br = Browser()
```

```python
# ignore robots.txt
br.set_handle_robots(False)

# fetch the stackoverflow.com url
br.open('http://www.stackoverflow.com')
```

```python
# now we need to get the search form, so we get the first form in page, which is the search form, in position 0
br.select_form(nr=0)

# now we set the value of the search field with the name property as 'q'
br.form['q'] = 'python'

# submit the request
response = br.submit()

# print the response, by calling the read() method
print response.read()
```
