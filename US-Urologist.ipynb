```python
from bs4 import BeautifulSoup
import urllib2
import time
import urlparse
import mechanize
import cookielib
import xlwt

```

```python
BASE_URL = "http://health.usnews.com"
wb = xlwt.Workbook()
ws = wb.add_sheet('Urologists')
ws.write(0,0, "Doctor's Name")
ws.write(0,1, "Location")
ws.write(0,2, "Hospital, Clinic")
ws.write(0,3, "Specialty Areas")
row = 1
```

```python
main_page = urlparse.urljoin(BASE_URL, "/doctors/location-index/urologists")
br = mechanize.Browser()
cj = cookielib.LWPCookieJar()
br.set_handle_robots(False)
br.set_cookiejar(cj)
br.addheaders = [('User-agent', 'Firefox')]
resp = br.open(main_page)
soup = BeautifulSoup(resp.get_data())
print soup
```

```python
states_urls = [BASE_URL + item.a['href'] for item in soup.findAll('li', 't-block')]
try:
    for state_url in states_urls:
        br_state = mechanize.Browser()
        cj = cookielib.LWPCookieJar()
        br_state.set_cookiejar(cj)
        br_state.set_handle_robots(False)
        br_state.addheaders = [('User-agent', 'Firefox')]
        resp = br_state.open(state_url)
        state_soup = BeautifulSoup(resp.get_data())
        items = state_soup.findAll('div', 'item')
        for item in items:
            city_urls = [urlparse.urljoin(BASE_URL, a['href']) for a in item.findAll('a')]
            for city_url in city_urls:
                br_city = mechanize.Browser()
                cj2 = cookielib.LWPCookieJar()
                br_city.set_cookiejar(cj2)
                br_city.set_handle_robots(False)
                br_city.addheaders = [('User-agent', 'Firefox')]
                resp = br_city.open(city_url)
                city_soup = BeautifulSoup(resp.get_data())
                city = city_soup.find('input', attrs={'id':'doctorSearch_location'})
                city = city['value']
                print city
                try:
                    result = city_soup.find('div', attrs={'id':'search-results'})
                    for tr in result.findAll('tr')[1:]:
                        name = tr.find('h4').find('a').get_text()
                        specialty = tr.find('div', 't').get_text()
                        hospitals = [li.a.get_text() for li in tr.findAll('li', 't-slack')]
                        print "\t", name, "\t", specialty
                        hosp = set()
                        for item in hospitals:
                            if item and item != "":
                                hosp.add(item)
                                
                    hospitals = ", ".join(hosp)
                    ws.write(row, 0, name)
                    ws.write(row, 1, city)
                    ws.write(row, 2, hospitals)
                    ws.write(row, 3, specialty)
                    row = row + 1
                except:
                    pass
                    
            break
    except:
wb.save("Urologists.xls")
```
